```python
import gzip
import json
import numpy as np
import pandas as po
from tqdm import tqdm
```

```python
with open('data/qar_products_Automotive.jsonl') as f:
    jl = list(f)
```

```python
df = po.DataFrame()
for i, json_str in tqdm(enumerate(jl)):
    result = json.loads(json_str)
    
    row = {}
    row['asin'] = result['asin']
    row['category'] = result['category']
    
    for question in result['questions']:
        
        questionText = question['questionText']
        questionType = question['questionType']
        
        for answer in question['answers']:
            answerText = answer['answerText']
            answerType = answer['answerType']
            answerHelpfulness = answer['helpful']
            
            if answerHelpfulness in [[0,0], [1,1]]:
                continue
            
            row['questionText'] = questionText
            row['questionType'] = questionType 
            row['answerText'] = answerText 
            row['answerType'] = answerType 
            row['answerHelpfulness'] = answerHelpfulness
            
            df = df.append(row, ignore_index = True)
```

```python
df
```

```python
df = df[['asin', 'category', 'questionText', 'questionType', 'answerText', 'answerType', 'answerHelpfulness']]
```

```python
df
```

```python
answerHelpfulness = np.array(df['answerHelpfulness'].to_list())
```

```python
df['Helpfulness_Target'] = np.where(np.logical_and(answerHelpfulness[:, 0] == answerHelpfulness[:, 1], answerHelpfulness[:, 1] >= 2), 1, 0)
```

```python
meta = gzip.open("data/meta_Automotive.json.gz","r")
```

```python
asins = df['asin'].to_list()
```

```python
df_meta = po.DataFrame()
for i in tqdm(meta):
    data = eval(i) 
    
    try:
        if data['asin'] not in asins:
            continue
    except:
        continue
    
    row = {}
    row['asin'] = data['asin']
    
    try:
        row['title'] = data['title']
    except:
        continue
        #row['title'] = ''
    
    try:
        row['description'] = data['description']
    except:
        #print('Product {} no description'.format(row))
        row['description'] = ''
        
    try:
        row['details'] = data['details']
    except:
        #print('Product {} no description'.format(row))
        row['details'] = ''
    
    df_meta = df_meta.append(row, ignore_index = True)
```
